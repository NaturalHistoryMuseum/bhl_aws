### Merge all cleaned OCR and try to find possible authors and any identified names

```python
bhl_file = '~/Downloads/BHL_data/page.txt'
bhl_page_df = load_tab_separated_file(bhl_file)
```

```python
# Step 0: Directory with OCR CSVs
input_dir = os.path.expanduser('~/Documents/bhl_aws/bhl_ocr')
all_files = [f for f in os.listdir(input_dir) if f.endswith('.csv')]

# Step 1: Load and concatenate all CSVs
merged_rows = []
for f in all_files:
    path = os.path.join(input_dir, f)
    
    # Extract BarCode and SequenceOrder from filename
    filename = os.path.splitext(f)[0]
    barcode, seq_order = filename.rsplit('_', 1)

    # Read CSV to get all columns temporarily
    df_raw = pd.read_csv(path, header=None)
    if df_raw.shape[1] < 3:
        continue  # skip malformed files
    
    # Extract only first, second, and last column
    df = pd.DataFrame()
    df["Author"] = df_raw.iloc[:, 0]
    df["Title"] = df_raw.iloc[:, 1]
    df["StartPageNumber"] = df_raw.iloc[:, -1]

    # Add metadata from filename
    df["BarCode"] = barcode
    df["SequenceOrder"] = int(seq_order)
    
    merged_rows.append(df)

# Combine all into one DataFrame
ocr_df = pd.concat(merged_rows, ignore_index=True)
```

```python
ocr_df
```

```python
# Step 2: Match BarCode → ItemID from bhl_item_df
ocr_df = ocr_df.merge(bhl_item_df[['ItemID', 'BarCode']], on='BarCode', how='left')

# Step 3: Use (ItemID, SequenceOrder) to get PageID from bhl_page_df
# ⚡ Efficient lookup via tuple index
bhl_page_df = bhl_page_df[['ItemID', 'SequenceOrder', 'PageID']].copy()
bhl_page_df['ItemID'] = bhl_page_df['ItemID'].astype(int)
bhl_page_df['SequenceOrder'] = bhl_page_df['SequenceOrder'].astype(int)

# Convert to dict for fast lookup
page_lookup = {
    (row.ItemID, row.SequenceOrder): row.PageID
    for row in bhl_page_df.itertuples(index=False)
}

# Map to PageID
ocr_df['PageID'] = ocr_df.apply(
    lambda row: page_lookup.get((row['ItemID'], row['SequenceOrder'])), axis=1
)
```

```python
ocr_df.to_csv('~/Downloads/350_ocr_df.csv',encoding='utf-8')
```

```python
# Filter bhl_name_df to only PageIDs of interest
valid_page_ids = ocr_df['PageID'].dropna().unique()
filtered_name_df = bhl_name_df[bhl_name_df['PageID'].isin(valid_page_ids)].copy()

# Prepare and clean NameConfirmed
filtered_name_df['NameConfirmed'] = filtered_name_df['NameConfirmed'].fillna('').astype(str)

# Group all NameConfirmed entries per PageID (in case multiple names per page)
name_by_page = (
    filtered_name_df.groupby('PageID', as_index=False)
    .agg({'NameConfirmed': lambda x: ' | '.join(x)})
)

# Step 4: Merge back to ocr_df
ocr_with_name = ocr_df.merge(name_by_page, on='PageID', how='left')
```

```python
import requests
from pygbif import species
```

```python
df = pd.read_csv('~/Documents/bhl_aws/BHL_preprocess/744_parts_processed.csv',encoding='utf-8')
```

```python
# Single full name string per entry
names = [
    "Beilschmiedia manantlanensis Cuevas, R | Cochrane, Theodore S.",
    "Quercus robur L."
]

for full_name in names:
    result = species.name_backbone(name=full_name)
    print({
        "input": full_name,
        "match": result.get("scientificName"),
        "canonical": result.get("canonicalName"),
        "authorship": result.get("authorship"),
        "confidence": result.get("confidence")
    })
```

```python
df_selected = df[[
    'PartID', 'ItemID', 'SequenceOrder',
    'Title', 'Date', 'StartPageID', 'SegmentUrl',
    'CreatorName', 'NameConfirmedFromPage'
]].copy()
```

```python
from tqdm import tqdm  # Optional for progress bar

tqdm.pandas()

def parse_and_match(row):
    name_list = []

    # 1. Split NameConfirmedFromPage
    if pd.isna(row['NameConfirmedFromPage']):
        return pd.Series([[], []], index=["MatchedNames", "Confidences"])
    
    confirmed_names = row['NameConfirmedFromPage'].split(" | ")
    creator = row['CreatorName'] if pd.notna(row['CreatorName']) else ""

    name_to_parse = [f"{name.strip()} {creator.strip()}" for name in confirmed_names if name.strip()]

    matched_names = []
    confidences = []

    for full_name in name_to_parse:
        try:
            result = species.name_backbone(name=full_name)
            confidence = result.get("confidence", 0)
            if confidence > 80:
                matched_names.append(result.get("scientificName"))
                confidences.append(confidence)
        except Exception as e:
            continue  # Skip failed lookups

    return pd.Series([matched_names, confidences], index=["MatchedNames", "Confidences"])
```

```python
# Apply to new DataFrame
df_selected[['MatchedNames', 'Confidences']] = df_selected.progress_apply(parse_and_match, axis=1)
```

```python
df_selected.to_csv('df_selected.csv',encoding='utf-8')
```

```python
def clean_and_sort_matches(row):
    names = row['MatchedNames']
    scores = row['Confidences']
    
    if not isinstance(names, list) or not isinstance(scores, list):
        return pd.Series([[], []], index=["MatchedNames", "Confidences"])
    
    # Remove None and keep valid pairs
    valid_pairs = [(name, score) for name, score in zip(names, scores) if name is not None]
    
    # Sort by confidence descending
    sorted_pairs = sorted(valid_pairs, key=lambda x: x[1], reverse=True)
    
    # Separate back into two lists
    cleaned_names = [name for name, _ in sorted_pairs]
    cleaned_scores = [score for _, score in sorted_pairs]
    
    return pd.Series([cleaned_names, cleaned_scores], index=["MatchedNames", "Confidences"])
```

```python
sorted_df = df_selected.copy()
sorted_df[['MatchedNames', 'Confidences']] = sorted_df.apply(clean_and_sort_matches, axis=1)
```

```python
def extract_best_match(row):
    names = row['MatchedNames']
    scores = row['Confidences']
    
    if not isinstance(names, list) or not isinstance(scores, list):
        return None

    for name, score in zip(names, scores):
        if score == 100:
            return name  # return the first match with confidence 100

    return None  # if no confidence == 100
```

```python
# Apply to sorted_df to create best_match column
sorted_df['best_match'] = sorted_df.apply(extract_best_match, axis=1)
```

```python
sorted_df['has_best_match'] = sorted_df['best_match'].notna()
```
